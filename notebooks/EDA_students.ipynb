```python
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------- CLEAN GENDER COLUMN ----------------

df['gender'] = (
    df['gender']
    .astype(str)
    .str.strip()
    .str.lower()
)

# Remove unwanted categories if any
df = df[df['gender'].isin(['male', 'female'])]

# ---------------- CREATE FIGURE ----------------

f, ax = plt.subplots(1, 2, figsize=(20, 10))

# ---------------- COUNTPLOT ----------------

sns.countplot(
    x='gender',
    data=df,
    palette='bright',
    ax=ax[0],
    saturation=0.95
)

# Add labels on bars
for container in ax[0].containers:
    ax[0].bar_label(
        container,
        color='black',
        size=15
    )

ax[0].set_title('Gender Countplot', fontsize=20)

# ---------------- PIE CHART ----------------

size = df['gender'].value_counts()

labels = size.index

colors = ['#ff4d4d', '#ff8000']

# explode length automatically matches categories
explode = [0.1] * len(size)

ax[1].pie(
    size,
    labels=labels,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    colors=colors
)

ax[1].set_title('Gender Pie Chart', fontsize=20)

# ---------------- FINAL LAYOUT ----------------

plt.tight_layout()

plt.show()
```

```python
import pandas as pd

# ---------------- CHECK DATA TYPES ----------------

print(df.dtypes)

# ---------------- CONVERT NUMERIC COLUMNS ----------------

score_cols = [
    'math_score',
    'reading_score',
    'writing_score',
    'science_score',
    'total_score'
]

for col in score_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# ---------------- CLEAN GENDER COLUMN ----------------

df['gender'] = (
    df['gender']
    .astype(str)
    .str.strip()
    .str.lower()
)

# ---------------- HANDLE MISSING VALUES ----------------

for col in score_cols:
    df[col] = df[col].fillna(df[col].mean())

# ---------------- GROUP BY GENDER ----------------

gender_group = df.groupby('gender')[score_cols].mean()

# ---------------- DISPLAY RESULT ----------------

print(gender_group)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------- CREATE AVERAGE COLUMN ----------------

df['average'] = (
    df['math_score'] +
    df['reading_score'] +
    df['writing_score'] +
    df['science_score']
) / 4

# ---------------- GROUP BY GENDER ----------------

gender_group = df.groupby('gender').mean(numeric_only=True)

# ---------------- PLOT DATA ----------------

plt.figure(figsize=(10, 8))

X = ['Total Average', 'Math Average']

female_scores = [
    gender_group.loc['female', 'average'],
    gender_group.loc['female', 'math_score']
]

male_scores = [
    gender_group.loc['male', 'average'],
    gender_group.loc['male', 'math_score']
]

X_axis = np.arange(len(X))

# Male Bars
plt.bar(
    X_axis - 0.2,
    male_scores,
    0.4,
    label='Male'
)

# Female Bars
plt.bar(
    X_axis + 0.2,
    female_scores,
    0.4,
    label='Female'
)

# Labels
plt.xticks(X_axis, X)

plt.ylabel("Marks")

plt.title(
    "Total Average vs Math Average Marks by Gender",
    fontweight='bold'
)

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------- GROUP DATA ----------------

Group_data2 = df.groupby('race_ethnicity')

# ---------------- CREATE SUBPLOTS ----------------

f, ax = plt.subplots(1, 3, figsize=(20, 8))

# ---------------- MATH SCORE ----------------

sns.barplot(
    x=Group_data2['math_score'].mean().index,
    y=Group_data2['math_score'].mean().values,
    palette='mako',
    ax=ax[0]
)

ax[0].set_title(
    'Math Score',
    color='#005ce6',
    size=20
)

# Add labels
for container in ax[0].containers:
    ax[0].bar_label(
        container,
        color='black',
        size=15
    )

# ---------------- READING SCORE ----------------

sns.barplot(
    x=Group_data2['reading_score'].mean().index,
    y=Group_data2['reading_score'].mean().values,
    palette='flare',
    ax=ax[1]
)

ax[1].set_title(
    'Reading Score',
    color='#005ce6',
    size=20
)

# Add labels
for container in ax[1].containers:
    ax[1].bar_label(
        container,
        color='black',
        size=15
    )

# ---------------- WRITING SCORE ----------------

sns.barplot(
    x=Group_data2['writing_score'].mean().index,
    y=Group_data2['writing_score'].mean().values,
    palette='coolwarm',
    ax=ax[2]
)

ax[2].set_title(
    'Writing Score',
    color='#005ce6',
    size=20
)

# Add labels
for container in ax[2].containers:
    ax[2].bar_label(
        container,
        color='black',
        size=15
    )

# ---------------- FINAL LAYOUT ----------------

plt.tight_layout()

plt.show()
```
